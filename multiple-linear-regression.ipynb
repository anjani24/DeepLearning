```python
path='/kaggle/input/multiple-linear-regression-dataset/multiple_linear_regression_dataset.csv'
```

```python
df= pd.read_csv(path)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= df[["age","experience"]]
X=np.array(X)
y=df[["income"]]
y=np.array(y)
```

```python
#train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)

X_train = (X_train - X_mean) / X_std
X_test  = (X_test - X_mean) / X_std
```

```python
class LinearReg:
    def __init__(self,learning_rate=0.0001, epochs=3000):
        self.lr= learning_rate
        self.epochs=  epochs
    def fit(self, X,y):
        n_samples, n_features= X.shape
        self.weight= np.zeros(n_features)
        self.bias=0
        
        for _ in range(self.epochs):
            for i in range (n_samples):
                y_pred=np.dot(X[i],self.weight)+self.bias
                error= y[i]-y_pred
                self.weight += self.lr* error* X[i]
                self.bias += self.lr * error
    def predict(self,X):
        return np.dot(X,self.weight)+self.bias
```

```python
model = LinearReg(
    learning_rate=0.001,
    epochs=3000
)

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def r2_score(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    return 1 - (ss_residual / ss_total)
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

```python
print("Weights:", model.weight)
print("Bias:", model.bias)

print("\nTest Predictions:", y_pred)
print("Actual Values:", y_test)

print("\nMean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))

rmse_value = rmse(y_test, y_pred)
print("RMSE:", rmse_value)
```
